```python
from pyspark.sql.types import LongType, IntegerType
```

```python
from pyspark.sql.types import StructType
```

```python
from pyspark.sql.types import StructField
```

```python
markSchema = StructType(fields=[
    StructField("user_id", IntegerType()),
    StructField("item_id", IntegerType()),
    StructField("rating", IntegerType()),
    StructField("timestamp", LongType())
])
```

```python
dfMark = spark.read.csv("/labs/laba01/ml-100k/u.data", sep="\t", schema=markSchema)
```

```python
dfMark.show(3)
```

```python
import pyspark.sql.functions as f
```

```python
dfMark.groupBy("item_id", "rating").agg(f.count("rating").alias("user_marks")).orderBy("rating").where("item_id = '302'").show()
```

```python
dfMark.registerTempTable("marks_table")
```

```python
spark.sql("select * from marks_table where item_id = 302 and rating = 2").show()
```

```python
dfMark.groupBy("rating").agg(f.count("rating").alias("user_marks")).orderBy("rating").show()
```

```python
spark.sql("select count(*) from marks_table where rating = 2").show()
```
